```python
import torchvision
from torchvision import transforms
import torch

datas=torchvision.datasets.CIFAR10(root='./datasets',transform=transforms.ToTensor(),train=True,download=True)   
```

```python
from torch.utils.data import DataLoader

loader= DataLoader(datas,batch_size=32,shuffle=True)
```

```python
import torch.nn as nn
from torch.nn import functional as F
from torch.nn import Conv2d

class myNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1=Conv2d(3,6,3)
    def forward(self,x):
        return self.conv1(x)
```

```python
from torch.utils.tensorboard import SummaryWriter

myNet=myNetwork()

writer=SummaryWriter('log3')

step=0
for data in loader:
    imgs,labels=data
    print(imgs.shape)
    output=myNet(imgs)
    print(output.shape)
    output=torch.reshape(output,(-1,3,30,30))
    writer.add_images('output1',output,global_step=step)
    step+=1
writer.close()
```
